```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl


# Create input variables
goout = ctrl.Antecedent(np.linspace (0, 5, 100), 'goout')
famrel = ctrl.Antecedent(np.linspace(0, 5, 100), 'famrel')
freetime = ctrl.Antecedent(np.linspace(0, 5, 100), 'freetime')
studytime = ctrl.Antecedent(np.linspace(0, 5, 100), 'studytime')
Pstatus = ctrl.Antecedent(np.linspace(0, 5, 100), 'Pstatus')

# Create output variable
health = ctrl.Consequent(np.linspace(0, 5, 100), 'health')

# Define membership functions for input variables
goout['very_less_goout_time'] = fuzz.gaussmf(goout.universe, 1, 0.6)
goout['less_goout_time'] = fuzz.gaussmf(goout.universe, 2, 0.6)
goout['average_goout_time'] = fuzz.gaussmf(goout.universe, 3, 0.6)
goout['good_goout_time'] = fuzz.gaussmf(goout.universe, 4, 0.6)
goout['max_goout_time'] = fuzz.gaussmf(goout.universe, 5, 0.6)

famrel['very_bad_relationship'] = fuzz.gaussmf(famrel.universe, 1, 0.6)
famrel['bad_relationship'] = fuzz.gaussmf(famrel.universe, 2, 0.6)
famrel['average_relationship'] = fuzz.gaussmf(famrel.universe, 3, 0.6)
famrel['good_relationship'] = fuzz.gaussmf(famrel.universe, 4, 0.6)
famrel['very_good_relationship'] = fuzz.gaussmf(famrel.universe, 5, 0.6)

freetime['very_less_freetime'] = fuzz.gaussmf(freetime.universe, 1, 0.6)
freetime['less_freetime'] = fuzz.gaussmf(freetime.universe, 2, 0.6)
freetime['average_freetime'] = fuzz.gaussmf(freetime.universe, 3, 0.6)
freetime['good_freetime'] = fuzz.gaussmf(freetime.universe, 4, 0.6)
freetime['max_freetime'] = fuzz.gaussmf(freetime.universe, 5, 0.6)

studytime['very_less_studytime'] = fuzz.gaussmf(studytime.universe, 1, 0.6)
studytime['less_studytime'] = fuzz.gaussmf(studytime.universe, 2, 0.6)
studytime['average_studytime'] = fuzz.gaussmf(studytime.universe, 3, 0.6)
studytime['good_studytime'] = fuzz.gaussmf(studytime.universe, 4, 0.6)
studytime['max_studytime'] = fuzz.gaussmf(studytime.universe, 5, 0.6)

Pstatus['together'] = fuzz.gaussmf(Pstatus.universe, 1, 0.6)
Pstatus['divorced'] = fuzz.gaussmf(Pstatus.universe, 0, 0.6)

# Define membership functions for output variable
health['very_bad'] = fuzz.gaussmf(health.universe, 1, 0.6)
health['bad'] = fuzz.gaussmf(health.universe, 2, 0.6)
health['average'] = fuzz.gaussmf(health.universe, 3, 0.6)
health['good'] = fuzz.gaussmf(health.universe, 4, 0.6)
health['very_good'] = fuzz.gaussmf(health.universe, 5, 0.6)


# Rule 1
rule1 = ctrl.Rule(goout['good_goout_time'] &
                  freetime['average_freetime'] &
                  studytime['less_studytime'] &
                  Pstatus['divorced'] &
                  famrel['very_good_relationship'],
                  health['bad'])


# Rule 2
rule2 = ctrl.Rule(goout['good_goout_time'] &
                  freetime['good_freetime'] &
                  studytime['less_studytime'] &
                  Pstatus['together'] &
                  famrel['bad_relationship'],
                  health['good'])


# Rule 3
rule3 = ctrl.Rule(goout['good_goout_time'] &
                  freetime['good_freetime'] &
                  studytime['average_studytime'] &
                  Pstatus['together'] &
                  famrel['average_relationship'],
                  health['good'])


# Rule 4
rule4 = ctrl.Rule(goout['good_goout_time'] &
                  freetime['good_freetime'] &
                  studytime['less_studytime'] &
                  Pstatus['together'] &
                  famrel['good_relationship'],
                  health['bad'])


# Rule 5
rule5 = ctrl.Rule(goout['average_goout_time'] &
                  freetime['average_freetime'] &
                  studytime['good_studytime'] &
                  Pstatus['divorced'] &
                  famrel['very_bad_relationship'],
                  health['average'])


# Rule 6
rule6 = ctrl.Rule(goout['average_goout_time'] &
                  freetime['very_less_freetime'] &
                  studytime['less_studytime'] &
                  Pstatus['together'] &
                  famrel['good_relationship'],
                  health['good'])

# Rule 7
rule7 = ctrl.Rule(goout['less_goout_time'] &
                  freetime['good_freetime'] &
                  studytime['good_studytime'] &
                  Pstatus['together'] &
                  famrel['good_relationship'],
                  health['average'])


# Rule 8
rule8 = ctrl.Rule(goout['average_goout_time'] &
                  freetime['average_freetime'] &
                  studytime['good_studytime'] &
                  Pstatus['together'] &
                  famrel['average_relationship'],
                  health['average'])


# Rule 9
rule9 = ctrl.Rule(goout['average_goout_time'] &
                  freetime['less_freetime'] &
                  studytime['good_studytime'] &
                  Pstatus['together'] &
                  famrel['very_good_relationship'],
                  health['average'])


# Rule 10
rule10 = ctrl.Rule(goout['very_less_goout_time'] &
                   freetime['max_freetime'] &
                   studytime['very_less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['average'])


# Rule 11
rule11 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['good_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['bad_relationship'],
                   health['average'])

# Rule 12
rule12 = ctrl.Rule(goout['less_goout_time'] &
                   freetime['less_freetime'] &
                   studytime['very_less_studytime'] &
                   Pstatus['together'] &
                   famrel['average_relationship'],
                   health['average'])


# Rule 13
rule13 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['very_good_relationship'],
                   health['good'])


# Rule 14
rule14 = ctrl.Rule(goout['less_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['good'])


# Rule 15
rule15 = ctrl.Rule(goout['good_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['bad'])


# Rule 16
rule16 = ctrl.Rule(goout['less_goout_time'] &
                   freetime['good_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['average'])


# Rule 17
rule17 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['average_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['good'])


# Rule 18
rule18 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['average_relationship'],
                   health['bad'])


# Rule 19
rule19 = ctrl.Rule(goout['good_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['good_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['good'])


# Rule 20
rule20 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['very_less_studytime'] &
                   Pstatus['together'] &
                   famrel['average_relationship'],
                   health['good'])


# Rule 21
rule21 = ctrl.Rule(goout['max_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['average_studytime'] &
                   Pstatus['together'] &
                   famrel['very_good_relationship'],
                   health['average'])


# Rule 22
rule22 = ctrl.Rule(goout['good_goout_time'] &
                   freetime['good_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['good'])


# Rule 23
rule23 = ctrl.Rule(goout['good_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['good'])


# Rule 24
rule24 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['very_less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['average'])


# Rule 25
rule25 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['less_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['very_good_relationship'],
                   health['bad'])


# Rule 26
rule26 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['good_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['average_relationship'],
                   health['average'])


# Rule 27
rule27 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['less_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['very_good_relationship'],
                   health['bad'])


# Rule 28
rule28 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['good_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['average_relationship'],
                   health['average'])


# Rule 29
rule29 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['less_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['average'])

# Rule 30
rule30 = ctrl.Rule(goout['good_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['good_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['good'])


# Rule 31
rule31 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['very_less_studytime'] &
                   Pstatus['together'] &
                   famrel['average_relationship'],
                   health['average'])


# Rule 32
rule32 = ctrl.Rule(goout['good_goout_time'] &
                   freetime['less_freetime'] &
                   studytime['very_less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['bad'])

# Rule 33
rule33 = ctrl.Rule(goout['less_goout_time'] &
                   freetime['max_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['average'])


# Rule 34
rule34 = ctrl.Rule(goout['max_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['together'] &
                   famrel['good_relationship'],
                   health['bad'])


# Rule 35
rule35 = ctrl.Rule(goout['average_goout_time'] &
                   freetime['good_freetime'] &
                   studytime['less_studytime'] &
                   Pstatus['divorced'] &
                   famrel['good_relationship'],
                   health['good'])


# Rule 36
rule36 = ctrl.Rule(goout['less_goout_time'] &
                   freetime['average_freetime'] &
                   studytime['very_less_studytime'] &
                   Pstatus['together'] &
                   famrel['very_good_relationship'],
                   health['average'])
health_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27, rule28, rule29, rule30, rule31, rule32, rule33, rule34, rule35, rule36 ])


health_sim = ctrl.ControlSystemSimulation(health_ctrl)
goout_values = data['goout'].values
studytime_values = data['studytime'].values
Pstatus_values = data['Pstatus'].values
freetime_values = data['freetime'].values
famrel_values = data['famrel'].values



expected_health_values = []
predicted_health_values = []
accuracy_count = 0
for i in range(len(data)):
    health_sim.input['goout'] = goout_values[i]
    health_sim.input['studytime'] = studytime_values[i]
    health_sim.input['freetime'] = freetime_values[i]
    health_sim.input['Pstatus'] = Pstatus_values[i]
    health_sim.input['famrel'] = famrel_values[i]
    
    data['health']
    health_sim.compute()

    health_value = health_sim.output['health']
    print("Input: goout={}, studytime={},  freetime={}, Pstatus={},famrel={}, ".format(
        goout_values[i], studytime_values[i], 
        freetime_values[i],Pstatus_values[i], famrel_values[i]))
    print('Actual_value :', data['health'][i])
    expected_health_values.append(data['health'][i])
    predicted_health_values.append(health_value)
    if round(health_value) == round(data['health'][i]):
        accuracy_count += 1
    # Print the output value for 'health'
    print("Predicted value Output: health={}".format(round(health_value)))
    print("-----------------------------------------")
    

print(accuracy_count)
print(accuracy_count/len(data),'%')
```
