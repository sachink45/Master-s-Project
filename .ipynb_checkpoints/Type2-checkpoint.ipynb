```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('F:/Semister_3/final_dataset.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
feat = ['goout', 'freetime','studytime','Pstatus', 'famrel', 'health']
df = df[feat]
data_A = df[50:200]
```

```python
from pyit2fls import IT2FLS, IT2FS_Gaussian_UncertStd, IT2FS_plot, \
                     min_t_norm, max_s_norm, TR_plot, crisp
from numpy import linspace


domain = linspace(0.0, 5.0, 100)

very_bad_relationship = IT2FS_Gaussian_UncertStd(domain, [1, 0.6, 0.4, 0.8])
bad_relationship = IT2FS_Gaussian_UncertStd(domain, [2, 0.6, 0.4, 0.8])
average_relationship = IT2FS_Gaussian_UncertStd(domain, [3, 0.6, 0.4, 0.8])
good_relationship = IT2FS_Gaussian_UncertStd(domain, [4, 0.6, 0.4, 0.8])
very_good_relationship = IT2FS_Gaussian_UncertStd(domain, [5, 0.6, 0.4, 0.8])

very_less_free_time = IT2FS_Gaussian_UncertStd(domain, [1, 0.6, 0.4, 0.8])
less_free_time = IT2FS_Gaussian_UncertStd(domain, [2, 0.6, 0.4, 0.8])
average_free_time = IT2FS_Gaussian_UncertStd(domain, [3, 0.6, 0.4, 0.8])
good_free_time = IT2FS_Gaussian_UncertStd(domain, [4, 0.6, 0.4, 0.8])
max_free_time = IT2FS_Gaussian_UncertStd(domain, [5, 0.6, 0.4, 0.8])


very_less_study_time = IT2FS_Gaussian_UncertStd(domain, [1, 0.6, 0.4, 0.8])
less_study_time = IT2FS_Gaussian_UncertStd(domain, [2, 0.6, 0.4, 0.8])
average_study_time = IT2FS_Gaussian_UncertStd(domain, [3, 0.6, 0.4, 0.8])
good_study_time = IT2FS_Gaussian_UncertStd(domain, [4, 0.6, 0.4, 0.8])
max_study_time = IT2FS_Gaussian_UncertStd(domain, [5, 0.6, 0.4, 0.8])


very_less_goout_time = IT2FS_Gaussian_UncertStd(domain, [1, 0.6, 0.4, 0.8])
less_goout_time = IT2FS_Gaussian_UncertStd(domain, [2, 0.6, 0.4, 0.8])
average_goout_time = IT2FS_Gaussian_UncertStd(domain, [3, 0.6, 0.4, 0.8])
good_goout_time = IT2FS_Gaussian_UncertStd(domain, [4, 0.6, 0.4, 0.8])
max_goout_time = IT2FS_Gaussian_UncertStd(domain, [5, 0.6, 0.4, 0.8])


together = IT2FS_Gaussian_UncertStd(domain, [1, 1, 0.5, 1])
divorced = IT2FS_Gaussian_UncertStd(domain, [0, 1, 0.5, 1])

very_bad = IT2FS_Gaussian_UncertStd(domain, [1, 0.6, 0.4, 0.8])
bad = IT2FS_Gaussian_UncertStd(domain, [2, 0.6, 0.4, 0.8])
average = IT2FS_Gaussian_UncertStd(domain, [3, 0.6, 0.4, 0.8])
good = IT2FS_Gaussian_UncertStd(domain, [4, 0.6, 0.4, 0.8])
very_good = IT2FS_Gaussian_UncertStd(domain, [5, 0.6, 0.4, 0.8])


myIT2FLS = IT2FLS()

# Add input variables to the IT2FLS system
myIT2FLS.add_input_variable("goout")
myIT2FLS.add_input_variable("family_relationship")
myIT2FLS.add_input_variable("free_time")
myIT2FLS.add_input_variable("study_time")
myIT2FLS.add_input_variable("parents_status")

# Add output variable to the IT2FLS system
myIT2FLS.add_output_variable("health")
```

```python
myIT2FLS.add_rule([("goout", good_goout_time), ("free_time", average_free_time),
                   ("study_time", less_study_time),("parents_status", divorced),
                   ("family_relationship", very_good_relationship)], [("health", bad)])
myIT2FLS.add_rule([("goout", good_goout_time), ("free_time", good_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", bad_relationship)], [("health", good)])
myIT2FLS.add_rule([("goout", good_goout_time), ("free_time", good_free_time),
                   ("study_time", average_study_time),("parents_status", together),
                   ("family_relationship", average_relationship)], [("health", good)])
myIT2FLS.add_rule([("goout", good_goout_time), ("free_time", good_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", bad)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", average_free_time),
                   ("study_time", good_study_time),("parents_status", divorced),
                   ("family_relationship", very_bad_relationship)], [("health", average)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", very_less_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", good)])
myIT2FLS.add_rule([("goout", less_goout_time), ("free_time", good_free_time),
                   ("study_time", good_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", average)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", average_free_time),
                   ("study_time", good_study_time),("parents_status", together),
                   ("family_relationship", average_relationship)], [("health", average)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", less_free_time),
                   ("study_time", good_study_time),("parents_status", together),
                   ("family_relationship", very_good_relationship)], [("health", average)])
myIT2FLS.add_rule([("goout", very_less_goout_time), ("free_time", max_free_time),
                   ("study_time", very_less_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", average)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", good_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", bad_relationship)], [("health", average)])
myIT2FLS.add_rule([("goout", less_goout_time), ("free_time", less_free_time),
                   ("study_time", very_less_study_time),("parents_status", together),
                   ("family_relationship", average_relationship)], [("health", average)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", average_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", very_good_relationship)], [("health", good)])
myIT2FLS.add_rule([("goout", less_goout_time), ("free_time", average_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", good)])
myIT2FLS.add_rule([("goout", good_goout_time), ("free_time", average_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", bad)])
myIT2FLS.add_rule([("goout", less_goout_time), ("free_time", good_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", average)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", average_free_time),
                   ("study_time", average_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", good)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", average_free_time),
                   ("study_time", less_study_time),("parents_status", together),
                   ("family_relationship", average_relationship)], [("health", bad)])
myIT2FLS.add_rule([("goout", good_goout_time), ("free_time", average_free_time),
                   ("study_time", good_study_time),("parents_status", together),
                   ("family_relationship", good_relationship)], [("health", good)])
myIT2FLS.add_rule([("goout", average_goout_time), ("free_time", average_free_time),
                   ("study_time", very_less_study_time),("parents_status", together),
                   ("family_relationship", average_relationship)], [("health", good)])
myIT2FLS.add_rule([("goout", max_goout_time), ("free_time", average_free_time),
                   ("study_time", average_study_time),("parents_status", together),
                   ("family_relationship", very_good_relationship)], [("health", average)])
```

```python
count = 0
for i in range(50,100):
    it2out, tr = myIT2FLS.evaluate({"goout": data_A['goout'][i], "free_time": data_A['freetime'][i],
                                    "study_time": data_A['studytime'][i], "parents_status": data_A['Pstatus'][i],
                                    "family_relationship": data_A['famrel'][i]
                                   }, min_t_norm, max_s_norm, domain, method="Centroid", algorithm="KM")

    # Obtain the crisp output value
    crisp_output = round(crisp(tr["health"]))

    print('mental_health_of_the_student:', crisp_output)
    print('Actual_value:', data_A['health'][i])
    if crisp_output == df['health'][i]:
        count = count + 1
    print()
print(count)
print((count/50)*100,'%')
```
